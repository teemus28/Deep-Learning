```python
import tensorflow
from tensorflow.keras.preprocessing.image import ImageDataGenerator, load_img, img_to_array # Import load_img , generate new image
```

```python
img = load_img('/content/cat_image.jpg', target_size=(200, 200))
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img)
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,         # angle of rotation , ±40 degrees
        width_shift_range=0.2,    # shifts image horizontally
        height_shift_range=0.2,   # shifts image vertically
        rescale=1./255,          # normalize
        shear_range=0.2,      # distortion of image
        zoom_range=1.6,    #zoom 1.6X
        horizontal_flip=True,  #flips image left-right(horizontally)
        fill_mode='nearest'    #nearest pixel value is used to fill newly creted pixel
        )
```

```python
print(type(img))
```

```python
img = img_to_array(img)   # coverting img to array
```

```python
img.shape
```

```python
input_batch = img.reshape(1, 200, 200, 3)    # 1 img of size (200, 200, 3)
```

```python
i=0

for output in datagen.flow(input_batch, batch_size=1, save_to_dir='/content/cat_new_image'):   # flow (for single image) , create new images
  i = i + 1
  if i == 10:   # breaks at 10
    break
```
