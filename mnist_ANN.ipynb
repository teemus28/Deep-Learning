```python
import pandas as pd
import tensorflow
from tensorflow import keras
from keras.datasets import mnist
```

```python
(X_train, y_train),( X_test, y_test) = mnist.load_data()
```

```python
X_train = X_train/255          #normalize between 0-1
X_test = X_test/255
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
model = Sequential()
model.add(Flatten(input_shape=(28,28)))
#model.add(Dense(32, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
# from tensorflow.keras.optimizers import SGD # Import the SGD optimizer   , 0.9221

# model.compile(loss='sparse_categorical_crossentropy',
#                 optimizer=SGD(momentum=0.9,nesterov=False), # Use SGD with momentum=0.9
#                 metrics=['accuracy'])
```

```python
#model.compile(loss='sparse_categorical_crossentropy',optimizer='RMSProp',metrics=['accuracy']) #0.9256
```

```python
model.compile(loss='sparse_categorical_crossentropy',optimizer='Adam',metrics=['accuracy'])    # accuracy = correct prediction/total prediction , 0.9241
```

```python
history = model.fit(X_train,y_train,epochs=10,validation_split=0.2)   # 20% validation data
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred.shape
```

```python
y_pred = y_pred.argmax(axis=1)
y_pred
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
```
